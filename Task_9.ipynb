## Sentiment Analysis

```python
# Model
mdlSnt = pipeline('sentiment-analysis')

# Parameters
#prmStatement = 'We are happy to go on vacation this spring break.'
prmStatement = ["We will be going camping to Kemer. I heard that Caretta Caretta eggs will be hatched, let's go watch them",
        "I am struggling because of the hot weather", "ohh the weather is very hot"]
# Predict
mdlSnt(prmStatement)
```

Let's use oxymoron to test it

```python
statement = ["Almost exactly.",
"Awfully good.",
"Deafening silence.",
"Freezer Burn.",
"Living dead.",
"Loyal opposition.",
"Old news.",
"Only choice."]
mdlSnt(statement)
```

Can perform sentiment analysis in other languages as well using `model = 'nlptown/bert-base-multilingual-uncased-sentiment'`

```python
# Model
mdlSnt = pipeline('sentiment-analysis', model = 'nlptown/bert-base-multilingual-uncased-sentiment')

# Parameters
prmStatement = 'film korkunç derecede iyiydi' #spanish for: the weather is very hot

# Predict
mdlSnt(prmStatement)
```

The result is in stars as opposed to score... this is as per the model definition.  See model details in HF for more details.

## Emotion Analysis:

```python
mdlEmotion = pipeline("text-classification", model="seara/rubert-tiny2-ru-go-emotions")

emotions = ["Her husband is so moody that she never knows if he will be happy or angry when she gets home from work.",
            "I trust that everything will work out in the end.",
            "They accepted that life was unpredictable and enjoyed the ride.",
            "The defendant showed no emotion when the verdict was read.",
            "I can't let my emotions interfere with a mission"]

mdlEmotion(emotions)
```

Let's mix it up by using a little complicated sentences


```python
emotions2 = ["I’m thrilled that I finally got the promotion, but now I have to work twice as hard to keep up with the demands.",
"Oh, great! Another meeting that could have been an email. Just what I needed today.",
"Winning the lottery was a blessing and a curse—now everyone expects me to solve their problems.",
"I’m happy to have received the gift, but I’m not sure why someone would think I’d want a pet snake.",
"Graduation day was full of joy, but the thought of leaving my friends behind made it hard to smile.",
"I’m hopeful that the new project will succeed, but given our track record, I can’t help but be a little skeptical.",
"It was a mistake to leave the company, but I’ve learned so much about myself since then.",
"I’m glad the exam is over, though I can’t shake the feeling that I could have done better."]

mdlEmotion(emotions2)
```

For instance, the last sentence (I’m glad the exam is over, though I can’t shake the feeling that I could have done better.) is not representing joy. Let's see what happens if we change the order of the sentence.

```python
trial= "Even though I can’t shake the feeling that I could have done better, I’m glad the exam is over."

mdlEmotion(trial)
```

as you can see the answers are not that satisfactory

So, we will use other models to answer the questions.

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

tokenizer = AutoTokenizer.from_pretrained("lserinol/bert-turkish-question-answering")
model = AutoModelForQuestionAnswering.from_pretrained("lserinol/bert-turkish-question-answering")
```

```python
text = r"""Antalya, Türkiye'nin güney sahilinde yer alan turistik bir şehir olup, Akdeniz Bölgesi'nde bulunmaktadır.
Şehir, özellikle yaz aylarında turist akınına uğrar. Antalya, tarihi dokusu, doğal güzellikleri ve mavi bayraklı plajlarıyla ünlüdür.
Şehirde, antik tiyatrolar, müzeler, doğal parklar ve şelaleler gibi birçok turistik cazibe merkezi bulunur.
Ayrıca, Antalya'nın merkezi ve çevresinde pek çok lüks otel, tatil köyü ve golf sahası yer alır.
Şehir, tarihi Kaleiçi bölgesi ve modern limanıyla da dikkat çeker.
Antalya, aynı zamanda önemli bir tarım bölgesidir ve tarım ürünleri ihracatında önemli bir role sahiptir.
Bölgedeki seralar, özellikle sebze ve meyve üretiminde büyük bir paya sahiptir.
Antalya'nın iklimi, tipik Akdeniz iklimidir; yazlar sıcak ve kurak, kışlar ise ılıman geçer."""

questions = ["Antalya hangi bölgede yer alır?",
"Antalya'nın hangi özellikleri turistik açıdan ünlüdür?",
"Antalya'nın iklimi nasıldır?",
"Antalya'da tarım hangi ürünler üzerinde yoğunlaşmıştır?"
]

for question in questions:
    inputs = tokenizer(question, text, add_special_tokens=True, return_tensors="pt")
    input_ids = inputs["input_ids"].tolist()[0]

    text_tokens = tokenizer.convert_ids_to_tokens(input_ids)
    outputs = model(**inputs) # Get the model outputs as a dictionary

    answer_start_scores = outputs.start_logits # Extract the start logits tensor
    answer_end_scores = outputs.end_logits # Extract the end logits tensor

    answer_start = torch.argmax(
        answer_start_scores
    )  # Get the most likely beginning of answer with the argmax of the score
    answer_end = torch.argmax(answer_end_scores) + 1  # Get the most likely end of answer with the argmax of the score

    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))

    print(f"Question: {question}")
    print(f"Answer: {answer}\n")
```

Again for the last question the answer is not that perfect but since the answer cannot be in the text it is enough.
