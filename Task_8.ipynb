## Build Recommendations

So, to be able to recommend jokes to users accoring to their preferences there are 3 types of **Recommer Systems:**
1. Content Based Filtering: if the same user previously was interested in a similar object the recommendations are made accordingly
2. Bundling Recommendation
3. Collobrative Filtering: groups the similar users, make the recommendations according to that clusters (the content is not important) (sort of matching users with similar users and making recommendations accordingly)


We will first start with Content Based Filtering which builds a recommender using content of the jokes.

### 1. Content Based Filtering

The aim of this recommer system is to determine how similar the descriptions are based on the terms used in the descriptions - while ignoring commonly used words.  Then recommend other jokes with similar descriptions. In order to do this, **TF-IDF Vectorization** is used.

For instance, a users rates jokes with the word 'lightbulb' in it highly he/she will be recommended other jokes with the same word in them.

#### Prepare data

```python
dfRatings.shape
```

```python
dfJokes.shape
```

```python
# Remove duplicates
dfJokes.drop_duplicates(subset ='JokeText', keep = 'first', inplace = True)
dfJokes.shape
```

so there were no duplicates of the jokes

and since there is no missing value we don't need to fill them with space

#### Build Model

```python
# Generate a matrix of common terms that show up in each movie

from sklearn.feature_extraction.text import TfidfVectorizer

mdlTfvJokes = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0.0, stop_words='english')
tfidf_matrix = mdlTfvJokes.fit_transform(dfJokes['JokeText'])
tfidf_matrix.shape
```

The similarity between any two Jokes let say (x) and (y) is defined as the **Cosine Similarity**: cosine(x,y)=x.y⊺||x||.||y||

Since we used the TF-IDF Vectorizer, calculating the Dot Product will directly give the Cosine Similarity Score.

```python
# Calculate cosine similarity between each pair of jokes as a function of the similarity of the common terms

from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

#### Predict

```python
# Prepare recommendation function
jokes = dfJokes['JokeText']
indices = pd.Series(dfJokes.index, index=dfJokes['JokeText'])

def get_recommendations(joke):
    idx = indices[joke]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    joke_indices = [i[0] for i in sim_scores]
    return jokes.iloc[joke_indices]
```

Now, we will built a model to recommend jokes according to sense of humor of a specific person.

For this, we will use a technique called Collaborative Filtering. For example, a person rated similarly with another person and their sense of humor can be considered as similar. So, this Technique help to recommend jokes with that idea.

### 2. Collaborative Filtering

The idea here is that a user x Jokes matrix is decomposed into a product of:

user x concept . concept x concept . concept x movie matrices.  

These can then be used to derive similarities between users.  This process is known as **Singular Value Decomposition**.

#### Prepare data

```python
dfRatings.head()
```

```python
dfJokeRatings = dfRatings.melt(id_vars=['JokeId'], var_name = 'User ID', value_name = 'Rating')
dfJokeRatings.head(10)
```

```python
dfJokeRatings.dropna(inplace=True)
dfJokeRatings.reset_index(drop=True, inplace=True)
```

#### Build Model

```python
# Prepare data into Surprise library format

!pip3 install scikit-surprise #or !conda install -c conda-forge scikit-surprise
from surprise import Dataset
from surprise import Reader
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(-10,10))
X = Dataset.load_from_df(dfJokeRatings[['User ID','JokeId', 'Rating']], reader)
X_train, X_test = train_test_split(X, test_size=.25)
```

```python
# Define SVD model

from surprise import SVD

mdlSvdRatings = SVD()
```

```python
# Fit SVD model

mdlSvdRatings.fit(X_train)
test_pred = mdlSvdRatings.test(X_test)
```

```python
# Evalute SVD accuracy

from surprise import accuracy

accuracy.rmse(test_pred)
```

```python
# Tune hyperparameters

from surprise.model_selection import GridSearchCV

param_grid = {'n_epochs': [5, 10, 15], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(X)

# best RMSE score
print(gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
# Cross-validate

from surprise.model_selection import cross_validate

cross_validate(mdlSvdRatings, X, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
